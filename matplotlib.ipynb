```python
import seaborn as sns
data = sns.load_dataset('tips')
data.head()
```

```python
plt.figure(figsize=(10,5))
plt.plot(data.tip,color='red',label='tips')
plt.plot(data.total_bill,color='green',label='total_bill')
plt.grid(linestyle='--',color='red',alpha=0.3)
plt.legend()
```

```python
#subplots
```

```python
plt.figure(figsize=(8,6))
plt.subplot(1,2,1)
plt.plot(data.tip,color='red',label='tips')
plt.xlabel=('x_axis')
plt.subplot(1,2,2)
plt.plot(data.total_bill,color='green',label='total_bill')
plt.ylabel=('total bill') 
plt.subplots_adjust(wspace=0.5 , hspace=0.3)
plt.show() 
```

```python
plt.figure(figsize=(10,8))
plt.subplot(3,3,2)
plt.plot(data.tip,color='red',label='tips')
plt.subplot(3,3,3)
plt.plot(data.total_bill,color='green',label='total_bill', linestyle='--' , marker='p')
plt.subplot(3,3,4)
plt.plot(data.day,color='blue',label='day')
plt.subplot(3,3,6)
plt.plot(data.sex,color='red',label='sex')
plt.subplot(3,3,7)
plt.plot(data.smoker, color='pink',label='smoker',linestyle='-.')
plt.subplot(3,3,8)
plt.plot(data.time, color='purple',label='time')
```

```python
x = np.random.randint(1, 100, 20)
bin_edges = [2,4,6,8,10,12,14,15]
plt.hist(x, color = 'green' , bins = bin_edges , alpha = 0.7,edgecolor='red',histtype='step',linestyle='--')
#histype : step,stepfilled,bar
plt.show()
```

```python
#piechart
```

```python
plt.figure(figsize = (5,5))
company = ['mercedes','bmw','tata','swift','nano']
revenue = [10,20,30,60,50]
colors = ['red','green','pink','blue','yellow']   
plt.pie(x = revenue , labels = company , colors = colors , autopct = '%.2f%%' , explode = [0,0.4,0,0,0.23] , textprops = {'color' : 'purple'})
plt.legend(loc = 'lower left')
plt.show()
```

```python
#boxplot
```

```python
x = [-120,10,20,30,40,50,60,100,200]
plt.boxplot(x, vert = False, medianprops={'color':'red' , 'linestyle':'--' , 'linewidth':3} , capprops={'color':'green'} , flierprops={'marker':'*' , 'markersize':20 , 'markerfacecolor':'red'})
plt.show()
```

```python
x = [10,20,30,40,5,60]
var = plt.violinplot(x, vert = False, showmeans=True , showmedians=True)
var['cmeans'].set_color('yellow')
var['cmedians'].set_color('green')
var['cbars'].set_color('red')
plt.show()
```
