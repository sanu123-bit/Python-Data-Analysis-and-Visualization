```python
#scatter plot
#hue is used to add third category style 4th col    and    legend is alredy created here by hue palette in mat first use label and then legend
# orient changes x to y and y to x
plt.figure(figsize=(6,4))
plt.title('Total Bill VS Tips')
palette={'Thur':'red', 'Fri':'yellow', 'Sat':'green', 'Sun':'Pink'}
sns.scatterplot(data=df , x='total_bill', y='tip', hue='day', palette=palette, style='sex', marker='*', s=100)  #s gives size 
plt.show()
```

```python
#barplot
#if we dont write data = df we have to write it manually then
#black is estimator
#dodge False gives the overlapping (comparison)
#apha gives transparency
#if we give x to y and y to x it will be vertical
palette={'Male':'green', 'Female':'pink'}
sns.barplot(x=df['day'], y=df['tip'], hue=df['sex'], order=['Sun', 'Thur', 'Fri', 'Sat'], hue_order=['Female', 'Male'], palette=palette, dodge=True,
            alpha=0.4)
```

```python
#hist plot
#it has continuous values  in x ( values can be numerics,float)
#kde density - which range has more data
#bins gves which data we want in particular range
#elemts giv
palette={'Male':'green', 'Female':'yellow'}
sns.histplot(data=df, x='tip',hue='sex', fill=False, palette=palette, kde=True, bins=10, element='step', stat='percent' #density, percent, count
            , hue_order=['Female', 'Male'], shrink=0.2)
```

```python
#kernel density estimator
#we can add line properties coz it is line based
sns.kdeplot(data=df, x='tip', hue='sex', linestyle='dashdot', fill=True, linewidth=4)
```

```python
#box plot
#no alpha
#means are traingles
#fill pours the colour
#width gives the size of the box
#numerical categorical in x num in y
palette={'Male':'blue', 'Female':'Pink'}
sns.boxplot(data=df, x='day', y='tip', palette=palette, hue='sex', showmeans=True, showfliers=True, whis=3, hue_order=['Female','Male'],
            meanprops={'marker':'p', 'markerfacecolor':'red','markeredgecolor':'red'},
            flierprops={'marker':'*'}, fill=False, width=0.5, fliersize=4, dodge=True,
            medianprops={'color':'yellow', 'linestyle':'-.', 'linewidth':2})
             
            
           
```

```python
#count plot
#kis basis pe classify so use hue
#no y coz it automatically counts the value
sns.set_style('whitegrid')  #dark
sns.countplot(data=df, x='day', hue='sex', palette={'Male':'cyan', 'Female':'orange'} , alpha=0.3, order=['Sat','Sun','Thur','Fri'])
#coolwarm, rocket husl,dark,Blues,Oranges  these are palettes
```

```python
#pairplot    gives access of particular value to multiple values
sns.pairplot(data=df)
```

```python
sns.pairplot(data=df, hue='sex', palette='Blues',kind='reg', markers='*', dropna=True)  #kind = reg ,scatter ,hist
```

```python
#violin plot  
# which place has more data saturation udhar line ko mota krega
#inner= sticks, box, quartile
#density_norms=area, count
sns.violinplot(data=df, x='day', y='tip', hue='sex', palette='Oranges', split=True, inner='sticks', linewidth=1, linestyle='-', hue_order=['Female','Male'],
               order=['Sat','Sun','Thur','Fri'], width=1, density_norm='area')
```
